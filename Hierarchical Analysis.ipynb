```python
# Import Dependencies
import matplotlib.pyplot as plt  
import pandas as pd  
%matplotlib inline
import numpy as np
```

```python
# 
enron_dataset = pd.read_csv('dataset/enron_dataset.csv')
display(enron_dataset.shape)
```

```python
data = enron_dataset.iloc[:, 3:5].values  
data
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))  
plt.title("Hierarchical Analysis of Enron POI")  
dend = shc.dendrogram(shc.linkage(data, method='ward')) 
plt.show()
```
